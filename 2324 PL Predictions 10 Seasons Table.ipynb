```python
def random_score(xGh, xGa):
  """
  Assigns a random score using `xGh` and `xGa`.

  Args:
    xGh: The predicted number of goals scored by the home team.
    xGa: The predicted number of goals scored by the away team.

  Returns:
    A random score for the home team and a random score for the away team.
  """

  home_score = np.random.poisson(xGh)
  away_score = np.random.poisson(xGa)

  return home_score, away_score
```

```python
# Convert the Home variable to a DataFrame
data['Home'] = pd.DataFrame(data['Home'])

# Convert the Away variable to a DataFrame
data['Away'] = pd.DataFrame(data['Away'])

# Replace the value 1 in the Home variable with the level Arsenal
data['Home'].replace(1, 'Arsenal', inplace=True)

# Replace the value 1 in the Away variable with the level Arsenal
data['Away'].replace(1, 'Arsenal', inplace=True)

# Create a Poisson regression model
home_model = smf.glm(formula="xGh ~ + Home + Away", data=data, family=sm.families.Poisson()).fit()

# Create a Poisson regression model for the away team
away_model = smf.glm(formula="xGa ~ + Home + Away", data=data, family=sm.families.Poisson()).fit()

# Predict the number of goals scored by the home team
predictionsH = home_model.predict(data)

# Predict the number of goals scored by the away team
predictionsA = away_model.predict(data)

# Add the predictions to the DataFrame
data['PredictedH'] = predictionsH.round(2)

# Add the predictions to the DataFrame
data['PredictedA'] = predictionsA.round(2)

# Create a table of the results
table = data[['Wk', 'Home', 'PredictedH', 'PredictedA', 'Away']]

#Optional if you want to print the predicted Home and Away xG (remove # below)
#print(table.to_string())

# Create a new column called home_score
data['home_score'] = [np.random.poisson(predictionsH[i]) for i in range(len(data))]

# Create a new column called away_score
data['away_score'] = [np.random.poisson(predictionsA[i]) for i in range(len(data))]

# Create a new column called HomeResults
data['HomeResults'] = [3 if home_score > away_score else 1 if home_score == away_score else 0 
                        for home_score, away_score in zip(data['home_score'], data['away_score'])]

data['AwayResults'] = [3 if away_score > home_score else 1 if away_score == home_score else 0 
                        for away_score, home_score in zip(data['away_score'], data['home_score'])]

# Create a table of the results
table = data[['Wk', 'Home', 'PredictedH', 'PredictedA', 'Away','HomeResults','AwayResults','home_score','away_score']]

# Optional if you want to Print the table to see the predicted scores
#print(table.to_string())

teams = data['Home'].unique()

results = {}
for team in teams:
    results[team] = data[data['Home'] == team]['HomeResults'].sum() + data[data['Away'] == team]['AwayResults'].sum()
    
# Create a dictionary to store the points tally
points_tally = {team: 0 for team in teams}

# Calculate the points for each team
Results1 = {}
for team in teams:
    Results1[team] = data[data['Home'] == team]['HomeResults'].sum() + data[data['Away'] == team]['AwayResults'].sum()

# Sort the teams based on points in descending order
sorted_teams = sorted(Results1.keys(), key=lambda x: results[x], reverse=True)

# Create a DataFrame for the Premier League table
premier_league_table = pd.DataFrame({'Team': sorted_teams, 'Results1': [Results1[team] for team in sorted_teams]})

df= premier_league_table

# Set the index to be the team names
premier_league_table.set_index('Team', inplace=True)

# Sort the table based on points in descending order
premier_league_table.sort_values('Results1', ascending=False, inplace=True)

# Print the Premier League table
print(premier_league_table)
```
